```python

# ðŸš€ Exploratory Data Analysis (EDA) on Titanic Dataset
# Tools Used: Pandas, Matplotlib, Seaborn, Plotly


# â­ Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# 1ï¸âƒ£ Load the Dataset

df = pd.read_csv("/content/Titanic-Dataset.csv")

# Display first few records
print("\nðŸ”¹ First 5 Rows:")
print(df.head())

# Basic structure of the dataset
print("\nðŸ”¹ Dataset Info:")
print(df.info())


# 2ï¸âƒ£ Summary Statistics


# Numerical summary (mean, std, min, max)
print("\nðŸ”¹ Summary Statistics (Numerical):")
print(df.describe())

# Categorical summary (unique, freq)
print("\nðŸ”¹ Summary Statistics (Categorical):")
print(df.describe(include='object'))

# Check missing values
print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())


# 3ï¸âƒ£ Histograms for All Numeric Columns

num_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(12, 10))
df[num_cols].hist(bins=30, figsize=(12, 10))
plt.suptitle("Histograms of Numeric Features")
plt.tight_layout()
plt.show()


# 4ï¸âƒ£ Boxplots for Outlier Detection

plt.figure(figsize=(12, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")

plt.tight_layout()
plt.show()

# 5ï¸âƒ£ Correlation Matrix + Heatmap

corr = df[num_cols].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()


# 6ï¸âƒ£ Pairplot for Relationship Between Features

sns.pairplot(df[num_cols])
plt.suptitle("Pairplot of Numeric Columns", y=1.02)
plt.show()

# 7ï¸âƒ£ Interactive Visuals Using Plotly


# Interactive Histogram of Age
fig = px.histogram(df, x='Age', title="Interactive Age Distribution")
fig.show()

# Interactive Scatterplot (Age vs Fare)
fig = px.scatter(df, x='Age', y='Fare', color='Survived',
                 title="Age vs Fare Colored by Survival")
fig.show()

# ðŸŽ‰ End of EDA Code

print("\nâœ” EDA Completed Successfully!")
```
